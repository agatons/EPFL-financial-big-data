```python
import pandas as pd
```

```python
df = pd.read_pickle('../data/raw/swedish_stocks')
df.head(2)
```

```python
tickers = df.columns.get_level_values(0).unique()
for ticker in tickers:
    curr_equ = df[ticker].dropna(how='all')
    if curr_equ.shape[0] == 0: # Dont save empty stocks
        continue
    if 'Adj Close' in curr_equ:
        curr_equ.drop(columns=["Adj Close"], inplace=True)
    assert curr_equ.isna().sum().sum() == 0 # Make sure any certain values are not NaN
    assert str(ticker)[len(ticker)-3:] == '.ST' # Make sure that we dont remove anything relevant
    curr_equ.index = pd.to_datetime(curr_equ.index)
    curr_equ = curr_equ.round(2) 
    curr_equ.to_csv(path_or_buf='../data/clean/swe_equ/' + str(ticker)[:len(ticker)-3] + '.csv')
```
