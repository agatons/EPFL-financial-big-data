```python
import pandas as pd
import numpy as np
import re
import os
```

```python
data = pd.read_excel('./data/raw/financials_raw.xlsx', index_col = 0)
```

```python
pd.DataFrame(data[data['Company name Latin alphabet'] == 'AB VOLVO'].columns)
data.rename(columns = {'Company name Latin alphabet' : 'Name'}, inplace = True)
data.drop('Consolidation code', axis = 1, inplace = True)
data
```

```python
nice_data = pd.DataFrame(columns = list(['Year']) + list(data.columns[0:5]) + list(data.columns[5:126:10]))
nice_data.columns
```

```python
for column in nice_data.columns:
    nice_data.rename(columns = {column : re.sub('\n.*', '', column)}, inplace = True)
```

```python
temp = nice_data
for year in range(2010,2020):
    temp.Year = year    
    temp.iloc[:, 1:6] = data.iloc[:, 0:5] 
    for cols in range(5, 125, 10):
        temp.iloc[:, int(6+(cols-5)/10)] = data.iloc[:, cols]
    nice_data = nice_data.append(temp)
```
